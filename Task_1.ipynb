* Calculate % growth of funding between year 2008 & 2009.
* If funding is positive, tag it as 1, if funding is negative tag it as 0. This is the target variable.

```python
import pandas as pd
```

```python
file_path = "C:/Users/mayank/Downloads/word_cloud/2010 Federal STEM Education Inventory Data Set - Copy.xls" 
df = pd.read_excel(file_path, header = 1)
df
```

```python
df.shape
```

```python
# Calculate the growth parameter based on funding growth
df['Growth'] = (df['C2) Funding FY2009'] - df['C1) Funding FY2008'])/df['C1) Funding FY2008'] * 100
    # Assign categorical binary label based on growth 
Tag =[]
val = float()
for val in df.Growth:
    Tag.append('0') if val < 0 else Tag.append('1')
df['Target_var'] = pd.Series(Tag)
```
