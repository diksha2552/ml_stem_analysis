2) Calculate mutual_info_score of target variable created in stage 1 & ALL non funding variables and share on a jupyter notebook.

```python
# Calculate the growth parameter based on funding growth
df['Growth'] = (df['C2) Funding FY2009'] - df['C1) Funding FY2008'])/df['C1) Funding FY2008'] * 100
# Assign categorical binary label based on growth 
Tag =[]
val = float()
for val in df.Growth:
    Tag.append('0') if val < 0 else Tag.append('1')
df['Target_var'] = pd.Series(Tag)
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
df['Agency'].as_matrix(columns=None)
labels_a = df['Agency'].as_matrix(columns=None)
labels_b = df['Target_var'].as_matrix(columns=None)
    # Mutual information
mi = mutual_info_score(labels_a, labels_b)
print(mi)
```

```python
df['B) Year Established'].as_matrix(columns=None)
labels_a = df['B) Year Established'].as_matrix(columns=None)
labels_b = df['Target_var'].as_matrix(columns=None)
    # Mutual information
mi = mutual_info_score(labels_a, labels_b)
print(mi)
```
